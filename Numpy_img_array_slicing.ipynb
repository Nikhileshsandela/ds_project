So let's do some masking on the photo

```
photo_masked = np.where(photo > 100, 255, 0) # replace with 255 where >100, & with 0 where <=100
```

we've taken a threshold of 100 and we've replaced everything that's over 100 with the value 255 and everything that's under 100 or 100 and under with zero

```
plt.imshow(photo_masked)
```

```
photo_masked = np.where(photo > 50, 255, 0)
```

```
plt.imshow(photo_masked)
```
